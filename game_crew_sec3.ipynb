````python
%%writefile main.py
### Paste your code below
import pygame
import sys
import random

# --- Game Constants ---
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 400
GROUND_HEIGHT = 50
GROUND_LEVEL = SCREEN_HEIGHT - GROUND_HEIGHT

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (0, 150, 0)
RED = (200, 0, 0)
BLUE = (0, 0, 200)
GRAY = (100, 100, 100)
SKY_COLOR = (135, 206, 235) # Light blue for sky
GROUND_COLOR = (128, 128, 0) # Brownish-yellow for ground

# Dino properties
DINO_WIDTH = 40
DINO_HEIGHT = 60
DINO_START_X = 50
DINO_START_Y = GROUND_LEVEL - DINO_HEIGHT

# Physics (values converted to be frame-rate independent, based on 60 FPS nominal)
# For example, 0.8 pixels/frame^2 at 60 FPS is 0.8 * 60 * 60 = 2880 pixels/second^2
GRAVITY_PPSS = 2880.0  # Pixels per second squared
# For example, -15 pixels/frame at 60 FPS is -15 * 60 = -900 pixels/second
JUMP_VELOCITY_PPS = -900.0 # Pixels per second
JUMP_BOOST_FACTOR = 0.5 # How much holding jump key extends jump (0.0 to 1.0, 1.0 means no gravity while key held)

# Game speed (values converted to be frame-rate independent)
# For example, 5 pixels/frame at 60 FPS is 5 * 60 = 300 pixels/second
INITIAL_GAME_SPEED_PPS = 300.0 # Pixels per second
SPEED_INCREASE_INTERVAL = 30 # seconds
# For example, 0.5 pixels/frame increase at 60 FPS is 0.5 * 60 = 30 pixels/second
SPEED_INCREASE_AMOUNT_PPS = 30.0 # Pixels per second

# Obstacle generation
INITIAL_OBSTACLE_SPAWN_MIN_TIME = 1.0 # seconds
INITIAL_OBSTACLE_SPAWN_MAX_TIME = 2.5 # seconds
PTERO_HEIGHTS = [GROUND_LEVEL - 40, GROUND_LEVEL - 80, GROUND_LEVEL - 120] # Possible Ptero Y positions

# --- Game Classes ---

class Dino:
    def __init__(self, x, y):
        self.rect = pygame.Rect(x, y, DINO_WIDTH, DINO_HEIGHT)
        self.color = BLUE
        self.y_velocity = 0 # Pixels per second
        self.is_jumping = False
        self.jump_key_held = False # True if jump key is currently pressed down

    def jump(self):
        if not self.is_jumping:
            self.is_jumping = True
            self.y_velocity = JUMP_VELOCITY_PPS

    def update(self, dt):
        # Apply gravity, potentially modified by jump boost
        current_gravity = GRAVITY_PPSS
        # If jumping, key held, and moving upwards, reduce gravity effect
        if self.is_jumping and self.jump_key_held and self.y_velocity < 0:
            current_gravity *= (1.0 - JUMP_BOOST_FACTOR) # e.g., if factor 0.5, gravity is halved

        self.y_velocity += current_gravity * dt
        self.rect.y += self.y_velocity * dt

        # Check for landing on the ground
        if self.rect.bottom >= GROUND_LEVEL:
            self.rect.bottom = GROUND_LEVEL
            self.is_jumping = False
            self.y_velocity = 0

    def draw(self, screen):
        pygame.draw.rect(screen, self.color, self.rect)

class Obstacle:
    def __init__(self, x, y, width, height, color):
        self.rect = pygame.Rect(x, y, width, height)
        self.color = color

    def update(self, game_speed_pps, dt):
        self.rect.x -= game_speed_pps * dt
        # Return True if the obstacle is off-screen to the left
        return self.rect.right < 0

    def draw(self, screen):
        pygame.draw.rect(screen, self.color, self.rect)

class Cactus(Obstacle):
    def __init__(self, x, ground_level):
        # Cactus is single-block height
        super().__init__(x, ground_level - 30, 20, 30, GREEN)

class Rock(Obstacle):
    def __init__(self, x, ground_level):
        # Rock is double-block height
        super().__init__(x, ground_level - 60, 40, 60, GRAY)

class Pterodactyl(Obstacle):
    def __init__(self, x, ground_level):
        # Pterodactyl appears at varying heights
        y_pos = random.choice(PTERO_HEIGHTS)
        super().__init__(x, y_pos, 50, 30, RED)

# --- Game Initialization ---
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Dino Dash")
clock = pygame.time.Clock()

# Fonts
font = pygame.font.Font(None, 36)
big_font = pygame.font.Font(None, 72)

# --- Game Variables (for reset) ---
def reset_game():
    global dino, obstacles, game_speed_pps, score, last_obstacle_spawn_time, \
           next_obstacle_spawn_time, speed_increase_timer, game_state, final_score, \
           obstacle_spawn_min_time, obstacle_spawn_max_time, ground_x1, ground_x2

    dino = Dino(DINO_START_X, DINO_START_Y)
    obstacles = []
    game_speed_pps = INITIAL_GAME_SPEED_PPS
    score = 0.0 # Use float for score as it accumulates dt
    last_obstacle_spawn_time = 0.0 # Time since last obstacle was spawned
    next_obstacle_spawn_time = random.uniform(INITIAL_OBSTACLE_SPAWN_MIN_TIME, INITIAL_OBSTACLE_SPAWN_MAX_TIME)
    speed_increase_timer = 0.0 # Timer for increasing game speed
    game_state = "running"
    final_score = 0 # To store score when game ends

    obstacle_spawn_min_time = INITIAL_OBSTACLE_SPAWN_MIN_TIME
    obstacle_spawn_max_time = INITIAL_OBSTACLE_SPAWN_MAX_TIME

    # Ground scrolling positions
    ground_x1 = 0
    ground_x2 = SCREEN_WIDTH

# Initial game setup
reset_game()

# --- Main Game Loop ---
running = True
while running:
    # Calculate delta time for frame-rate independent movement
    # dt is time in seconds since last frame
    dt = clock.tick(60) / 1000.0 # Max 60 FPS

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:
            if game_state == "running":
                if event.key == pygame.K_SPACE or event.key == pygame.K_UP:
                    dino.jump()
                    dino.jump_key_held = True # Mark jump key as held
            elif game_state == "game_over":
                # Restart the game if any key is pressed
                reset_game()

        if event.type == pygame.KEYUP:
            if game_state == "running":
                if event.key == pygame.K_SPACE or event.key == pygame.K_UP:
                    dino.jump_key_held = False # Release jump key

    # --- Game Logic (only if game is running) ---
    if game_state == "running":
        # Update Dino
        dino.update(dt)

        # Scroll ground (frame-rate independent)
        ground_x1 -= game_speed_pps * dt
        ground_x2 -= game_speed_pps * dt
        # Reset ground segments when they move off-screen
        if ground_x1 < -SCREEN_WIDTH:
            ground_x1 = SCREEN_WIDTH
        if ground_x2 < -SCREEN_WIDTH:
            ground_x2 = SCREEN_WIDTH

        # Obstacle Generation
        last_obstacle_spawn_time += dt
        if last_obstacle_spawn_time >= next_obstacle_spawn_time:
            # Randomly choose obstacle type, making pterodactyl less frequent
            # Weights: Cactus (3x chance), Rock (3x chance), Pterodactyl (1x chance)
            obstacle_class = random.choices(
                population=[Cactus, Rock, Pterodactyl],
                weights=[3, 3, 1], # Pterodactyls are correctly less frequent here
                k=1
            )[0]

            obstacles.append(obstacle_class(SCREEN_WIDTH, GROUND_LEVEL))

            last_obstacle_spawn_time = 0
            # Set next spawn time with current difficulty parameters
            next_obstacle_spawn_time = random.uniform(obstacle_spawn_min_time, obstacle_spawn_max_time)

        # Update and remove off-screen obstacles
        obstacles_to_keep = []
        for obstacle in obstacles:
            if not obstacle.update(game_speed_pps, dt): # If obstacle is NOT off-screen
                obstacles_to_keep.append(obstacle)
        obstacles = obstacles_to_keep

        # Collision Detection
        for obstacle in obstacles:
            if dino.rect.colliderect(obstacle.rect):
                game_state = "game_over"
                final_score = int(score) # Store final score
                break # Exit loop once collision is found

        # Score Update (1 point per second survived)
        score += dt

        # Speed Increase
        speed_increase_timer += dt
        if speed_increase_timer >= SPEED_INCREASE_INTERVAL:
            game_speed_pps += SPEED_INCREASE_AMOUNT_PPS
            speed_increase_timer = 0
            # Gradually reduce min/max spawn times to increase difficulty
            obstacle_spawn_min_time = max(0.5, obstacle_spawn_min_time * 0.9)
            obstacle_spawn_max_time = max(1.0, obstacle_spawn_max_time * 0.9)


    # --- Drawing ---
    screen.fill(SKY_COLOR) # Fill background with sky color

    # Draw Ground
    pygame.draw.rect(screen, GROUND_COLOR, (ground_x1, GROUND_LEVEL, SCREEN_WIDTH, GROUND_HEIGHT))
    pygame.draw.rect(screen, GROUND_COLOR, (ground_x2, GROUND_LEVEL, SCREEN_WIDTH, GROUND_HEIGHT))

    # Draw Dino
    dino.draw(screen)

    # Draw Obstacles
    for obstacle in obstacles:
        obstacle.draw(screen)

    # Draw Score
    score_text = font.render(f"Score: {int(score)}", True, BLACK)
    screen.blit(score_text, (10, 10))

    # Draw Game Over screen
    if game_state == "game_over":
        game_over_text = big_font.render("GAME OVER", True, BLACK)
        final_score_text = font.render(f"Final Score: {final_score}", True, BLACK)
        restart_text = font.render("Press any key to restart", True, BLACK)

        # Center the text on the screen
        screen.blit(game_over_text, (SCREEN_WIDTH // 2 - game_over_text.get_width() // 2, SCREEN_HEIGHT // 2 - 50))
        screen.blit(final_score_text, (SCREEN_WIDTH // 2 - final_score_text.get_width() // 2, SCREEN_HEIGHT // 2 + 10))
        screen.blit(restart_text, (SCREEN_WIDTH // 2 - restart_text.get_width() // 2, SCREEN_HEIGHT // 2 + 60))

    # Update the display
    pygame.display.flip()

# --- Game Exit ---
pygame.quit()
sys.exit()
````
